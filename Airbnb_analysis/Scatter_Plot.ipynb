# Predictions vs. Actual Price

<br> For each city, after performing the analysis, a csv file was exported that contains the predictions and actual pricec for the test set. Here, we load the data and compare the predictions for different cities to evaluate the workflow. 

<br> Loading necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<br> List of the cities

```python
city_names = ['Montreal', 'New_Brunswick', 'Quebec_City', 'Toronto', 'Vancouver', 'Victoria']
```

<br> Simple function to load the csv file. The <i>(city name).csv</i> file contains two columns, the preditions and actual price of the listings

```python
def load_pred(city):
    df = pd.read_csv((f'{city}_pred.csv'))
    return df
```

<br> Comparing the prediction results for all cities

```python
fig, ax = plt.subplots(figsize=(10, 7))

plt.plot([0,5],[0,5], color='black')
for city in city_names:
    df = load_pred(city)
    sns.scatterplot(data=df, x=df.y_test, y=df.predictions, alpha=0.8, label='%s' % city)
#     sns.kdeplot(data=df, x=df.y_test, y=df.predictions, levels=2, fill=True, alpha=0.2)
plt.legend()
ax.set(xlabel='Actual Price', ylabel='Estimated Price', xlim=(1,4.5), ylim=(1,4.5));
xy_range=np.arange(1, 5, step=1)
ax.set_xticks(xy_range)
ax.set_xticklabels(10**xy_range)
ax.set_yticks(xy_range)
ax.set_yticklabels(10**xy_range)
ax.grid(True)
```
